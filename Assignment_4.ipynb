```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df['Cabin'] = df['Cabin'].bfill()
```

```python
df.dropna(inplace = True)
```
